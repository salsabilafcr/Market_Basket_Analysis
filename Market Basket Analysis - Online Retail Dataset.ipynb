# Data Cleaning

```python
df_clean = df.copy()
```

```python
# create date column
df_clean['date'] = pd.to_datetime(df_clean['order_date']).dt.date.astype('datetime64[ns]')
df_clean.info()
```

```python
# remove all rows without customer_id
df_clean = df_clean[~df_clean['customer_id'].isna()]
df_clean.info()
```

```python
# convert customer_id to string
df_clean['customer_id'] = df_clean['customer_id'].astype(str)
df_clean.info()
```

```python
# make all product_names lowercase
df_clean['product_name'] = df_clean['product_name'].str.lower()
df_clean.head(3)
```

```python
# remove all rows with product_code or product_name contains 'test'
df_clean = df_clean[~(df_clean['product_code'].str.lower().str.contains('test')) | 
                    ~(df_clean['product_name'].str.contains('test'))]
df_clean
```

```python
# remove rows with cancelled status, i.e. those whose order_id begins with 'C'
df_clean = df_clean[df_clean['order_id'].str[:1] !='C']
df_clean.head(3)
```

```python
# change the negative quantity value to positive because the negative value only indicates 
# that the order is canceled.
df_clean['quantity'] = df_clean['quantity'].abs()
df_clean.sample(3)
```

```python
# remove rows with negative price
df_clean = df_clean[~(df_clean['price']<0)]
df_clean.sample(3)
```

```python
# creates an amount value, which is the multiplication of quantity and price
df_clean['amount'] = df_clean['quantity'] * df_clean['price']
df_clean.head(3)
```

```python
# replace the product_name of a product_code that has multiple product_names 
# with one of its most frequently occurring product_names
most_freq_product_name = df_clean.groupby(['product_code','product_name'], as_index=False).agg(order_cnt=('order_id','nunique')).sort_values(['product_code','order_cnt'], ascending=[True,False])
most_freq_product_name['rank'] = most_freq_product_name.groupby('product_code')['order_cnt'].rank(method='first', ascending=False)
most_freq_product_name = most_freq_product_name[most_freq_product_name['rank']==1].drop(columns=['order_cnt','rank'])
df_clean = df_clean.merge(most_freq_product_name.rename(columns={'product_name':'most_freq_product_name'}), how='left', on='product_code')
df_clean['product_name'] = df_clean['most_freq_product_name']
df_clean = df_clean.drop(columns='most_freq_product_name')
df_clean.sample(3)
```

```python
# remove outlier
from scipy import stats
df_clean = df_clean[(np.abs(stats.zscore(df_clean[['quantity','amount']]))<3).all(axis=1)]
df_clean = df_clean.reset_index(drop=True)
df_clean.head(3)
```

# Preparing Basket Data

```python
basket = pd.pivot_table(df_clean, index='order_id', columns='product_name', values='product_code',
                       aggfunc='nunique', fill_value=0)
basket
```

```python
basket.info()
```

First, create a basket dataframe, to find out what products have been sold.

```python
def encode(x):
    if x == 0:
        return False
    if x > 0:
        return True
basket_encode = basket.applymap(encode)
basket_encode
```

Encode DataFrame basket with value True for all values above 0 and False for all values 0.

```python
basket_filter = basket_encode[(basket_encode>0).sum(axis=1)>1]
basket_filter
```

Take a transaction with more than one unique product. Because the objective of this analysis is to find out the relationship between one product and another in a transaction. Then, the code will take the rows that have a value greater than 0 or True and sum them to the right (row-wise), to find out how many total orders there are per each transaction.

# Applying Apriori Alogarithm

```python
pip install mlxtend
```

```python
from mlxtend.frequent_patterns import apriori

frequent_itemset = apriori(basket_filter, min_support=.01, use_colnames=True).sort_values('support', ascending=False).reset_index(drop=True)
frequent_itemset['product_cnt'] = frequent_itemset['itemsets'].apply(lambda x: len(x))
frequent_itemset
```

List frequent itemsets (a collection of products that are often purchased).

```python
from mlxtend.frequent_patterns import association_rules

product_association = association_rules(frequent_itemset, metric='confidence', 
                                        min_threshold=.7).sort_values(['support','confidence'], 
                                                                      ascending=[False,False]).reset_index(drop=True)
product_association
```

# Analysis Results

From the market basket analysis, we can conclude that:
- The product Red Hanging Heart T-Light Holder was purchased in 5.9% of the total transactions.
- Once a buyer purchases Red Hanging Heart T-Light Holder, 72% of them will purchase White Hanging Heart T-Light Holder as well.
- There is a strong relationship between the Red Hanging Heart T-Ligh Holder and the White Hanging Heart T-Ligh Holder, with the chance of both products being purchased together in a single transaction being four times greater than the chance of both products being purchased individually.
- Based on this insight, the White Hanging Heart T-Ligh Holder product can be placed near/next to the Red Hanging Heart T-Ligh Holder or both products can be sold in bundles.
